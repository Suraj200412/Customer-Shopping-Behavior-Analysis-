Treating null vaalue

## Size

```python
df['Size'].unique()
```

```python
df.head()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.groupby('Item Purchased')['Size'].unique()
```

```python
#treating null values and replacing all null values in the electronc categori
df.loc[df['Category']=='Electronics','Size']='Not Applicable' 
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#treating null values and replacing all null values in the Accessories categori
df.loc[df['Category']=='Accessories','Size']='Free Size' 
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#treating null values and replacing all null values in the Footwear categori
df.loc[df['Category']=='Footwear','Size']='Not Available' 
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#treating null values and replacing all null values in the Clothing categori
df.loc[(df["Category"]=="Clothing") & (df["Size"].isnull()),"Size"] = df[df["Category"]=="Clothing"]["Size"].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.isnull().sum()
```

```python
# Remove spaces
df['Size'] = df['Size'].str.strip()

# Convert to uppercase
df['Size'] = df['Size'].str.upper()

# Replace full names with short form
df['Size'] = df['Size'].replace({
    'SMALL': 'S',
    'MEDIUM': 'M',
    'LARGE': 'L',
    'EXTRA LARGE': 'XL',
    'X-LARGE': 'XL'
})

# Fill null values
df['Size'] = df['Size'].fillna('UNKNOWN')
```
